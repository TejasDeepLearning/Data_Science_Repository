# Visualizing The Gender Gap In College Degrees

In this project, we'll visualize the gender gap across college degrees awarded to both men and women. We'll generate line charts to compare across all the degree categories, ie: STEM, liberal arts and other categories. In the last step of this project, we'll explore how to export the final diagram we create as an image file.

Let's import our standard libraries for reading data into dataframes and visualizing them. We then print the first few rows of the data to get an understanding of what the data is about.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/Users/Tejas/csv_files/percent-bachelors-degrees-women-usa.csv')
data.head()
```

This is a relatively small project. We're only focusing on how we can improve our plots as a whole when looking at multiple columns of data, and how we can present our plots to a large audience by including colors that people who are color blind can see, removing non-data-ink like excess tick values and bottom labels, cleaning up our charts by removing chart junk and so on. I'd encourage reading about visualizations from Edward Tufte, he talks more about cleaning up and presenting graphs to people in detail.

```python
fig = plt.figure(figsize=(20, 24))

stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)

for sp in range(0, 18, 3):
    cats_index = int(sp/3)
    ax = fig.add_subplot(6, 3, sp+1)
    ax.tick_params(left=False, bottom=False, labelbottom=False)
    ax.plot(data['Year'], data[stem_cats[cats_index]], c=cb_dark_blue, linewidth=4)
    ax.plot(data['Year'], 100 - data[stem_cats[cats_index]], c=cb_orange, linewidth=4)
    ax.set_title(stem_cats[cats_index])
    for key, spine in ax.spines.items():
        spine.set_visible(False)
        
    ax.set_yticks([0, 100])
    ax.axhline(50, c='lightgrey', alpha=0.5)
    
    if cats_index == 0:
        ax.text(2003, 80, 'Women')
        ax.text(2005, 28, 'Men')
    elif cats_index == 5:
        ax.text(2005, 87, 'Men')
        ax.text(2003, 7, 'Women')  
        ax.tick_params(labelbottom=True)

        
for sp in range(1, 16, 3):
    cats_index = int(sp/3)
    ax = fig.add_subplot(6, 3, sp+1)
    ax.tick_params(left=False, bottom=False, labelbottom=False)
    ax.plot(data['Year'], data[lib_arts_cats[cats_index]], c=cb_dark_blue, linewidth=4)
    ax.plot(data['Year'], 100 - data[lib_arts_cats[cats_index]], c=cb_orange, linewidth=4)
    ax.set_title(lib_arts_cats[cats_index])
    for key, spine in ax.spines.items():
        spine.set_visible(False)
        
    ax.set_yticks([0, 100])
    ax.axhline(50, c='lightgrey', alpha=0.5)
    
    if cats_index == 0:
        ax.text(2003, 75, 'Women')
        ax.text(2005, 36, 'Men')
    if cats_index == 4:
        ax.tick_params(labelbottom=True)

for sp in range(2, 20, 3):
    cats_index = int(sp/3)
    ax = fig.add_subplot(6, 3, sp+1)
    ax.tick_params(left=False, bottom=False, labelbottom=False)
    ax.plot(data['Year'], data[other_cats[cats_index]], c=cb_dark_blue, linewidth=4)
    ax.plot(data['Year'], 100 - data[other_cats[cats_index]], c=cb_orange, linewidth=4)
    ax.set_title(other_cats[cats_index])
    for key, spine in ax.spines.items():
        spine.set_visible(False)
        
    ax.set_yticks([0, 100])
    ax.axhline(50, c='lightgrey', alpha=0.5)
    
    if cats_index == 0:
        ax.text(2003, 76, 'Women')
        ax.text(2005, 20, 'Men')
        
    elif cats_index == 5:
        ax.text(2005, 35, 'Women')
        ax.text(2005, 65, 'Men')
        ax.tick_params(labelbottom=True)

plt.savefig('gender_gaps.png')
plt.show()
```
